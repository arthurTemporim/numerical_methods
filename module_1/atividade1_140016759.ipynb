```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.exp(np.sin(x)) + np.sin(np.exp(x))
```

```python
def f_differentiate(x):
    return np.exp(x) * np.cos(np.exp(x)) + np.exp(np.sin(x)) * np.cos(x)
```

```python
def imprime_grafico(a, b, num, f):
    x = np.linspace(a,  b,  num)
    print(x)
    y = f(x)
    plt.plot(x,y)
```

```python
imprime_grafico(-2,2,500,f)
```

```python
x  = np.linspace(-2, 2, 500)
```

```python
y = f(x)
```

```python
plt.plot(x,y)
```

```python
x2 = np.linspace(-2,  2,  500)
```

```python
dy = f_differentiate(x)
```

```python
plt.plot(x2,dy)
```

```python
def f_prog(f, x, h):
    return (f(x+h)-f(x))/h
```

```python
dy2 = f_prog(f, x, 1)
c = dy-dy2
np.mean(c)
```

```python
dy2 = f_prog(f, x, 0.1)
c = dy-dy2
np.mean(c)
```

```python
dy2 = f_prog(f, x, 0.01)
c = dy-dy2
np.mean(c)
```

```python
dy2 = f_prog(f, x, 1)
c  =  (dy-dy2)/dy2 
np.mean(c)
```

```python
dy2 = f_prog(f, x, 0.1)
c  =  (dy-dy2)/dy2 
np.mean(c)
```

```python
dy2 = f_prog(f, x, 0.01)
c  =  (dy-dy2)/dy2 
np.mean(c)
```
