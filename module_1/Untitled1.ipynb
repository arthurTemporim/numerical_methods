```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imprime_grafico(a=0, b=0, num=500, f=None):
    x = np.linspace(a, b, num)
    y = f(x)
    plt.plot(x,y)
```

```python
def f(x):
    return x+1
def f2(x):
    return np.log(x**2)
```

```python
imprime_grafico(b=30, f=f)
```

```python
imprime_grafico(0, 10, 200, f2)
```
